## Predicting Whether a cancer is benign or malignant

The data is obtained from [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29)

### The motive behind the project:
According to [Center for Disease Control and Prevention](https://gis.cdc.gov/Cancer/USCS/DataViz.html):
- Cancer is the second leading cause of death in the United States, exceeded only by heart disease
- One of every four deaths in the United States is due to cancer
- In 2017, over 250,520 new cases of female breast cancer were reported in the United States, and 42,000 women died of this cancer
- For every 100,000 women, 125 new Female Breast Cancer cases were reported and 20 women died of this cancer


Analyzing the breast cancer data set and using it to predict whether a cancer is benign or malignant could help greatly in implementing early detection, treatment, and prevention strategies which will lead to the reduction of number of deaths and the overall cost

### Data Analysis

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, KFold, RandomizedSearchCV
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.svm import SVC 
%matplotlib inline
sns.set (color_codes=True)
```

```python
# load the data 
BC = pd.read_csv('data.csv')
```

```python
#Lets look at features of data
BC.head()
```

**Attribute Information:**

- ID number (We do not need this column so we will drop it)
- Diagnosis (M = malignant, B = benign) (This is our target column: label column)

The rest of the attributs (feature columns) consist of real numbers that are computed for each cell:
- radius (mean of distances from center to points on the perimeter)
- texture (standard deviation of gray-scale values)
- perimeter
- area
- smoothness (local variation in radius lengths)
- compactness (perimeter^2 / area - 1.0)
- concavity (severity of concave portions of the contour)
- concave points (number of concave portions of the contour)
- symmetry
- fractal dimension ("coastline approximation" - 1)

```python
## exploring the missing values
null_values = BC.isnull().sum()
print(null_values)
```

```python
# the columns (ID and Unnamed: 32) are not needed so I will drop them 
# We will split the data: y includes our labels and x includes our features
no_need = ['Unnamed: 32','id']
BC = BC.drop(no_need,axis = 1 )
BC['diagnosis'] = BC['diagnosis'].map({'M': 1, 'B':0})
y = BC["diagnosis"]       # M or B 
x = BC.drop("diagnosis",axis = 1 )
x.head()
```

## Examining correlations between varaibles and label

```python
correlation = BC.corr()

fig, ax = plt.subplots (figsize = (25, 25))
sns.heatmap(correlation, annot=True, linecolor='Black', linewidths=1)
```

**There is a strong correlation between malignancy and (radius, number of concave points, perimeter, and area)**

That is expected because these features were modeled in such way that higher values are typically associated with malignancy

```python
def plot_violinplot(feat_list):
    scaler = StandardScaler()
    feat_scaled = pd.DataFrame(scaler.fit_transform(BC[feat_list]),columns=feat_list, index = BC.index)
    data = pd.concat([BC['diagnosis'],feat_scaled],axis=1)
    df_melt = pd.melt(frame=data, value_vars=feat_list, id_vars=['diagnosis'])
    fig, ax = plt.subplots(1, 1, figsize = (10, 10))
    sns.violinplot(x="variable",y="value",hue = "diagnosis",data=df_melt,split = True, inner="quart",palette='Set2').set_title('Distribution of features among malignant and benign tumours'.format(feat_list))
    plt.xticks(rotation=45)
    L=plt.legend()
    L.get_texts()[0].set_text('Benign')
    L.get_texts()[1].set_text('Malignant')
```

```python
plot_violinplot(means)
```

```python
plot_violinplot(se)
```

```python
plot_violinplot(worst)
```

Generally speaking, for the benign median is lower than the malignant median for almost all features, which makes sense, because features were modeled such that higher values are typically associated with malignancy.

Also, the difference in distributions between the benign and malignant cells is expected since higher values indicate malignancy. The number of feature variables are widely distributed and dramatically vary from cell to cell. 

Finally, almost all of the variables show a right skewed distribution which is expected in medical data.

## Feature Selection

The dataset consists of 30 feature and one label. Some of these variables are highly correlated which will cause problems resulting in a decrease in prediction accuracy by the algorithms. 

There is a positive correlation between:
- radius_mean, perimeter_mean, area_mean
- Compactness_mean, concavity_mean and concave points_mean 
- radius_se , perimeter_se , area_se
- radius_worst, perimeter_worst, area_worst

I will keep one of each group   

```python
list_to_drop = ['radius_mean','perimeter_mean',
                'radius_worst','perimeter_worst','area_worst','texture_worst',
                'compactness_mean','concave points_mean',
                'radius_se','perimeter_se',
                'compactness_worst','concave points_worst','compactness_se','concave points_se']

x1 = x.drop(list_to_drop,axis = 1)
print(x1.shape)
x1.head()
print(y.shape)
```

## Machine Learning (SVM)

```python
# train and split the data 
X_train, X_test, y_train, y_test = train_test_split(x1,
                                                    y, test_size=0.30, random_state=42)
```

### SVM hyperparameter tuning, prediction, and accuracy:

I will use randomized search cv to find the best parameters for the SVM model

```python
svm_pram_dist = {"kernel":["linear","rbf"],
                 "gamma":["auto",'scale'],
                "C":[0.01,0.1,1,10,100]
                }

svm = SVC()

svm_cv = RandomizedSearchCV(svm,svm_pram_dist,cv=10)

svm_cv.fit(X_train,y_train)
grid_predictions = grid_predictions = svm_cv.predict(X_test) 
accuracy = svm_cv.score(X_test,y_test) 

print("Tuned svm Parameters:{}".format(svm_cv.best_params_))
print("The model has an accuracy of: {}".format(accuracy))
```

```python
svm.fit(X_train,y_train)
```

## Model performance assesment 

```python
## confusion Matrix 

cm = confusion_matrix(y_test, grid_predictions)

plt.figure(figsize=(5,5))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True);
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(accuracy)
plt.title(all_sample_title, size = 10);
```

```python
print(classification_report(y_test, grid_predictions))
```
